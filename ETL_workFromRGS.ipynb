```python
import pandas as pd
data=pd.read_excel("financial_sample.xlsx")
data
```

```python
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')

# Print cleaned column names

data
#print(data.columns)
```

```python
new_dataset = data[['segment','country','product', 'discount_band', 'units_sold','sale_price','sales','date','month_name', 'year']]
new_dataset.to_csv("new_dataset.csv", index=False)
new_dataset.head()
```

```python
df_renamed = new_dataset.rename(columns={
    "segment": "customer_name",
    "country": "city",
    "product": "product_name",
    "discount_band":"category",
    "units_sold":"quantity",
    "sale_price":"price",
    "sales":"total_sales",
    "month_name":"month" 
})
# Display updated column names
df_renamed
#print(df_renamed.columns)
```

```python
df_renamed['price'] = df_renamed['price'].astype(str).str.replace(r'\$', '', regex=True)
df_renamed['total_sales'] = df_renamed['total_sales'].astype(str).str.replace(r'\$', '', regex=True)
```

```python
df_renamed["quantity"] = pd.to_numeric(df_renamed["quantity"], errors="coerce")
df_renamed["price"] = pd.to_numeric(df_renamed["price"], errors="coerce")
df_renamed
```

```python
df_renamed["total_sales"] = df_renamed["quantity"] * df_renamed["price"]
df_renamed["date"] = pd.to_datetime(df_renamed["date"])
df_renamed
```

```python
customer_data = df_renamed[['customer_name','city']]
customer_data.to_csv("customer.csv", index=False)
customer_data
```

```python
date_data = df_renamed[['date','month','year']]
date_data.to_csv("date.csv", index=False)
date_data
```

```python
product_data = df_renamed[['product_name','category']]
product_data.to_csv("product.csv", index=False)
product_data
```

```python
import pandas as pd
import mysql.connector
import numpy as np

# Connect to MySQL
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='',
    database='finance'
)
cursor = conn.cursor()

# Check and enable foreign key constraints
cursor.execute("SHOW VARIABLES LIKE 'foreign_key_checks';")
result = cursor.fetchone()
print(f"Foreign Key Checks Enabled: {result[1]}")

if result[1] == '0':
    cursor.execute("SET FOREIGN_KEY_CHECKS = 1;")
    print("Foreign key checks enabled.")

# Load CSV files into MySQL tables
csv_files = {"product.csv": "product", "customer.csv": "customer", "date.csv": "dates", "sales.csv": "sales"}

for file, table in csv_files.items():
    df = pd.read_csv(file)

    # Replace NaN values with None (NULL in MySQL)
    df = df.replace({np.nan: None})

    for _, row in df.iterrows():
        columns = ', '.join(df.columns)
        placeholders = ', '.join(['%s'] * len(row))
        query = f"INSERT INTO {table} ({columns}) VALUES ({placeholders})"
        
        # Convert row to tuple and replace NaN values with None
        values = tuple(None if pd.isna(value) else value for value in row)

        try:
            cursor.execute(query, values)
        except mysql.connector.Error as err:
            print(f"❌ Error inserting into {table}: {err}")

    conn.commit()
    print(f"✅ Imported {file} into {table}")

cursor.close()
conn.close()
```
