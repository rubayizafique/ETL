# Transform Data

```python
New_Product = df1[['Make', 'Model']]
New_Product = New_Product.rename(columns={'Make': 'product_name', 'Model': 'category'})
New_Product
```

```python
New_Customer = df[['Segment','Country']]
New_Customer = New_Customer.rename(columns={'Segment': 'Customer_name', 'Country': 'City'})
New_Customer
```

```python
New_Date = df[['Month Number','Month Name','Year']]
New_Date = New_Date.rename(columns={'Month Number': 'Date', 'Month Name': 'Month', 'Year': 'Year'})
New_Date 
```

```python
sales = df[['Gross Sales','Sale Price']]
New_sales = sales.rename(columns={'Gross Sales': 'Quantity', 'Sale Price': 'Unit Price'})
New_sales
```

```python
New_sales["total_sales"] = New_sales["Quantity"] * New_sales["Unit Price"]
New_sales
```

```python
date_cols = New_Date[['Date']]
customer_cols = New_Customer[['Customer_name']]
product_cols = New_Product[['product_name']]
sales_cols = New_sales[['total_sales']]
```

```python
concant = pd.concat([date_cols, customer_cols, product_cols, sales_cols], axis=1, join='inner')
concant
```

```python
Sales_data =concant
Sales_data
```
